### 여러 개의 CSV파일을 읽어서 요약하기

```python
import glob
import os

all_files = glob.glob(os.path.join('sales*'))
all_files
```

```python
all_data=[]
for file in all_files:
    data=pd.read_csv(file)
    all_data.append(data)
    
all_data_concat=pd.concat(all_data,axis=0, ignore_index=True)
all_data_concat
```

```python
all_data=[]
for file in all_files:
    data=pd.read_csv(file)
    
    total_sales=pd.DataFrame([float(str(value).strip('$').replace(',',''))
                             for value in data['Sale Amount']]).sum()
    
    average_sales=pd.DataFrame([float(str(value).strip('$').replace(',',''))
                             for value in data['Sale Amount']]).mean()
    
    summary={'파일이름':os.path.basename(file),
            '총매출':total_sales,
            '평균매출':average_sales}
    
    all_data.append(pd.DataFrame(summary))
    
data_concat=pd.concat(all_data,axis=0,ignore_index=True)
data_concat
```
