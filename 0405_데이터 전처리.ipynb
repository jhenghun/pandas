### 매핑을 통해 데이터 변형하기 - Feature Engineering

```python
data=pd.DataFrame({'food':['bacon','pulled pork','bacon',
                           'Pastrami','corned beef','Bacon',
                           'pastrami','honey ham','nova lox'],
                   'ounces':[4,3,12,6,7,5,8,6,4]})
data
```

```python
meat_to_animal={'bacon':'pig',
               'pulled pork':'pig',
               'pastrami':'cow',
               'corned beef':'cow',
               'honey ham':'pig',
               'nova lox':'salmon'}
```

```python
data['food']=data.food.str.lower()
data
```

```python
data['animal']=data.food.map(meat_to_animal)
data
```

### 개별화와 양자화
- 연속형 데이터는 종종 개별로 분할하거나 분석을 위해 그룹별로 나누기도 함
- 수업에 참여하는 학생들의 데이터가 있고, 이를 나이대에 따라 분류한다고 가정해보자

```python
ages=[20,55,67,43,1,5,18,29,34,46,42,53,59,67,60,20,30]
```

```python
rule=[0,18,40,60,100]
```

```python
cats=pd.cut(ages,rule,labels=['미성년자','청년','중년','노년'])
```

```python
cats.codes
```

```python
cats.categories
```

```python
cats.value_counts()
```

```python
a = pd.cut(ages,4)
```

```python
a.value_counts()
```

```python
data=np.random.randn(1000)
cats = pd.qcut(data, 4)
cats
```

```python
cats.value_counts()
```
