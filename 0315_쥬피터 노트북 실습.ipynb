# 쥬피터 노트북 소개
## 쥬피터 단축키
- m: 메모셀
- y: 코드셀
- a: 위에 셀 추가
- b: 아래에 셀 추가
- shift+enter: 셀 실행
- x: 셀 삭제

## 셀 종류
- 코드셀, 메모셀
- 초록색 모드: 코드/문서를 작성할 수 있는 상태(Enter)
- 파란색 모드: 단축키를 입력할 수 있는 상태(Esc)
